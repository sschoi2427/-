```python
# ◀ 1. 좌측 시작 동그라미 버튼 누르기 - 잠시 뒤 지표가 추출됩니다 - 무시 경고가 뜨면 무시하고 진행하기
# ▼ 2. 맨 아래에 결과가 나옵니다.

# 코드 저장방법(선택)
# 1. 상단 바 - 파일 - 드라이브에 사본저장
# 2. 구글 드라이브 안에 Colab Notebook 폴더 안에 저장 파일이 생깁니다.

import time
import requests
from bs4 import BeautifulSoup

# 구글에서 주가 가져오기
def get_stock_html(url, name='vix'):
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    stock_today = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.VfPpkd-WsjYwc.VfPpkd-WsjYwc-OWXEXe-INsAgc.KC1dQ.Usd1Ac.AaN0Dd.QZMA8b > c-wiz > div > div:nth-of-type(1) > div > div.rPF6Lc > div > div:nth-of-type(1) > div > span > div > div')[0].text.replace(',','').replace('$','')
    stock_yesday = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.eYanAe > div:nth-of-type(2) > div')[0].text.replace(',','').replace('$','')
                               
    stock_today = round(float(stock_today),4)
    stock_yesday = round(float(stock_yesday),4)
    
    stock_per = round( ((float(stock_today) - float(stock_yesday)) / float(stock_yesday))*100, 4)
    
    if name == 'dji':
        stock_name = "다우 "
        stock_today = round(stock_today)
        stock_per = round(stock_per,2)
    elif name == 'us500':
        stock_name = "S&P500 "
        stock_today = round(stock_today)
        stock_per = round(stock_per,2)
    elif name == 'ixic':
        stock_name = "나스닥 "
        stock_today = round(stock_today)
        stock_per = round(stock_per,2)
    elif name == 'rus':
        stock_name = "러셀 "
        stock_today = round(stock_today)
        stock_per = round(stock_per,2)
    elif name == 'dbc':
        stock_name = "원자재 "
        stock_today = round(stock_today,2)
        stock_per = round(stock_per,2)
    elif name == 'vix':
        stock_today = round(float(stock_today),2)
        stock_name = "VIX "
    elif name == '10y':
        stock_today = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.VfPpkd-WsjYwc.VfPpkd-WsjYwc-OWXEXe-INsAgc.KC1dQ.Usd1Ac.AaN0Dd.QZMA8b > c-wiz > div > div:nth-of-type(1) > div > div.rPF6Lc > div > div:nth-of-type(1) > div > span > div > div')[0].text.replace(',','')
        stock_yesday = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.eYanAe > div:nth-of-type(2) > div')[0].text.replace(',','')
        stock_today = round(float(stock_today)/10, 4)
        stock_yesday = round(float(stock_yesday)/10, 4)
        stock_per = round( ((float(stock_today) - float(stock_yesday)) / float(stock_yesday))*100, 4)
        stock_name = "T10Y "
    elif name == '5y':
        stock_today = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.VfPpkd-WsjYwc.VfPpkd-WsjYwc-OWXEXe-INsAgc.KC1dQ.Usd1Ac.AaN0Dd.QZMA8b > c-wiz > div > div:nth-of-type(1) > div > div.rPF6Lc > div > div:nth-of-type(1) > div > span > div > div')[0].text.replace(',','')
        stock_yesday = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.eYanAe > div:nth-of-type(2) > div')[0].text.replace(',','')
        stock_today = round(float(stock_today)/10, 4)
        stock_yesday = round(float(stock_yesday)/10, 4)
        stock_per = round( ((float(stock_today) - float(stock_yesday)) / float(stock_yesday))*100, 4)
        stock_name = "T30Y5YS "
    elif name == '30y':
        stock_today = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.VfPpkd-WsjYwc.VfPpkd-WsjYwc-OWXEXe-INsAgc.KC1dQ.Usd1Ac.AaN0Dd.QZMA8b > c-wiz > div > div:nth-of-type(1) > div > div.rPF6Lc > div > div:nth-of-type(1) > div > span > div > div')[0].text.replace(',','')
        stock_yesday = soup.select('#yDmH0d > c-wiz.zQTmif.SSPGKf.u5wqUe > div > div.e1AOyf > div > div > main > div.Gfxi4 > div.eYanAe > div:nth-of-type(2) > div')[0].text.replace(',','')
        stock_today = round(float(stock_today)/10, 4)
        stock_yesday = round(float(stock_yesday)/10, 4)
        stock_per = round( ((float(stock_today) - float(stock_yesday)) / float(stock_yesday))*100, 4)
        stock_name = "T30Y5YS "
    else:
        pass
    
    if stock_per > 0:
        stock_per = "+"+str(round(stock_per,2)) + "%"
    else:
        stock_per = str(round(stock_per,2))+ "%"
        
    
    return stock_name, str(stock_today), stock_per, stock_yesday
 
# 링크
dow = 'https://www.google.com/finance/quote/.DJI:INDEXDJX'
sp = "https://www.google.com/finance/quote/.INX:INDEXSP"
nas = "https://www.google.com/finance/quote/.IXIC:INDEXNASDAQ"
ru = "https://www.google.com/finance/quote/RUT:INDEXRUSSELL"
comd = "https://www.google.com/finance/quote/DBC:NYSEARCA"
vx = "https://www.google.com/finance/quote/VIX:INDEXCBOE"
t10 = "https://www.google.com/finance/quote/TNX:INDEXCBOE"
#bit = "https://www.google.com/finance/quote/BTC-USD"
t5 = "https://www.google.com/finance/quote/FVX:INDEXCBOE"
t30 = "https://www.google.com/finance/quote/TYX:INDEXCBOE"

# 실행
re_dji = get_stock_html(dow, 'dji')
re_us = get_stock_html(sp, 'us500')
re_ixic = get_stock_html(nas, 'ixic')
re_rus = get_stock_html(ru, 'rus')
re_dbc = get_stock_html(comd, 'dbc')
re_vix = get_stock_html(vx, 'vix')
re_t10 = get_stock_html(t10, '10y')
re_t5 = get_stock_html(t5, '5y')
re_t30 = get_stock_html(t30, '30y')
print('-'*70)

# 시간
!pip install pytz
import pytz
from datetime import datetime
NY_tz = pytz.timezone('America/New_York')
now_time = datetime.now(NY_tz)
print('-'*70)
print("뉴욕시간 :", now_time.strftime("20%y년 %m월 %d일 %H시 %M분 %S초"))
print()

# 글자 붙이기
dji = re_dji[0] + re_dji[1] +"("+ re_dji[2] +"),"
us = re_us[0] + re_us[1] +"("+ re_us[2] +"),"
ixic = re_ixic[0] + re_ixic[1] +"("+ re_ixic[2] +"),"
rus = re_rus[0] + re_rus[1] +"("+ re_rus[2] +"),"
dbc = re_dbc[0] + re_dbc[1] +"("+ re_dbc[2] +"),"
vix = re_vix[0] + re_vix[1] +"("+ re_vix[2] +"),"
t10 = re_t10[0] + re_t10[1] +"("+ re_t10[2] +")"

yesday_30y5ys = round(float(re_t30[3])-float(re_t5[3]),4)
today_30y5ys = round(float(re_t30[1])-float(re_t5[1]),4)
per_30y5ys = round((today_30y5ys - yesday_30y5ys) / yesday_30y5ys*100, 2)
if per_30y5ys > 0:
    per_30y5ys = "+"+str(round(per_30y5ys,2)) + "%"
else:
    per_30y5ys = str(round(per_30y5ys,2))+ "%"
t30y5ys = re_t5[0] + str(today_30y5ys) +"("+ str(per_30y5ys) +")"

total = ixic + " " + rus + " " + dbc + " " + vix + " " + t10 + ", " + t30y5ys

# 출력하기
print(dji)
print(us)
print(total)

# ※ 설명
# ※ 미국주식 매크로 지표 추출 코드입니다
# ※ 다우, S&P500, 나스닥, 러셀, 원자재, VIX, 10년물채권금리, 30Y5YS 장단기금리차가 추출됩니다.
```
